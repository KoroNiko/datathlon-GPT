```python
df = Artist[['id','birthplace','name']].dropna(subset=['birthplace'])
df = pd.merge(df, Artwork, how='inner', left_on = 'id', right_on = 'artist')
df = df[['id_x','id_y','birthplace','name_x','name_y']]
df.rename(columns={"id_x": "artist_id", "id_y": "artwork_id", "birthplace": "place_id",
                  "name_x":"artist_name", "name_y":"painting_name"}, inplace=True)
df = pd.merge(df, Place, how='inner', left_on = 'place_id', right_on = 'id')
df.drop(['parent','id'], axis=1, inplace=True)
df.rename(columns={"name": "place_name"}, inplace=True)
place_artist_artwork = df
place_artist_artwork
```
